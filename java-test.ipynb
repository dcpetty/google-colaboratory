```python
#!/usr/bin/env python3
# https://stackoverflow.com/a/51775653
# https://chatgpt.com/share/6856de83-9570-8011-abec-4eee1aa17ce3
import os   # importing os to set environment variable

print(f"realpath('.') is: '{os.path.realpath('.')}'")

def install_java():
    # !apt update
    # !apt-cache search openjdk
    # !update-alternatives --display java
    # !update-alternatives --display javac

    # Install openjdk-8.
    print(f"!apt-get install -y openjdk-8-jdk-headless")
    !apt-get install -y openjdk-8-jdk-headless -qq > /dev/null  # install openjdk
    jdk = '/usr/lib/jvm/java-8-openjdk-amd64'
    # Link java & Javac at priority > 1111.
    !update-alternatives --install /usr/bin/java java "{jdk}/bin/java" 11111
    !update-alternatives --install /usr/bin/javac javac "{jdk}/bin/javac" 11111
    os.environ["JAVA_HOME"] = "{jdk}"                           # set environment variable
    !java -version                                              # check java version
install_java()
```
