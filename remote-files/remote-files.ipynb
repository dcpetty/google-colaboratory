# `remote-files`

This [Google Colab Notebook](https://colab.research.google.com/) includes access to remote files both from a repository and from a local drive.

```python
#!/usr/bin/env python3
#
# Download remote files, mount Google Drive, and set local values.
#
from os.path import join, realpath
content = realpath('.')
gdrive = join(content, 'gdrive')
notebooks = join(gdrive, 'MyDrive/Colab Notebooks')
repo = 'remote-files'   # repo path within the Drive/Colab Notebooks directory
repo_path = join(notebooks, repo)

# https://colab.research.google.com/github/AllenDowney/ThinkPython/blob/v3/chapters/chap04.ipynb
from os.path import basename, exists
def download(url):
    filename = basename(url)
    if not exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print("Downloaded " + str(local))
    return filename

# https://colab.research.google.com/notebooks/io.ipynb#scrollTo=u22w3BFiOveA
from google.colab import drive
drive.mount(gdrive)
# %cd repo_path   # cd so . is repo_path

download('https://raw.githubusercontent.com/dcpetty/google-colab/remote-files/github.txt')

# Set global variables
verbose = False # whether to print data
```
